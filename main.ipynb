```python
from datetime import datetime

import pandas as pd
import json
```

```python
with open('dataset_fb.json') as file:
    dataset = json.load(file)

with open('kasthuri.json') as file:
    kasthuri = json.load(file)

dataset = [*dataset, *kasthuri]

with open('dataset.json', 'w') as file:
    json.dump(dataset, file)
```

```python
with open('dataset.json', 'r') as file:
    dataset = json.load(file)

df_data = {
    "User": [],
    "Timestamp": [],
    "Text": []
}

for post in dataset:
    user = post['user']
    df_data["User"].append(user['name'])
    df_data["Timestamp"].append(post['timestamp'])
    df_data["Text"].append(post['text'])

df = pd.DataFrame(df_data)
df['Timestamp'] = df["Timestamp"].apply(datetime.fromtimestamp)
```

```python
df.groupby('User').describe()
```

```python
from collections import Counter
import malaya

preprocesing = malaya.preprocessing.preprocessing()
tokenizer = malaya.tokenizer.Tokenizer()
fasttext = malaya.language_detection.fasttext()

counter = Counter()

def normalize_text(text):
    tokens = preprocesing.process(text)
    return " ".join(tokens)

df = df[df.Text != ""]
df['Language'] = fasttext.predict(df['Text'].values.tolist())
```

```python
df.groupby('Language').describe()
```

```python
corpus = df.Text.values.tolist()

vectorizer = malaya.transformer.load(model="electra")
lda2vec = malaya.topic_model.transformer.attention(
    corpus, n_topics=10, vectorizer=vectorizer
)

lda2vec.get_topics(10)
```
